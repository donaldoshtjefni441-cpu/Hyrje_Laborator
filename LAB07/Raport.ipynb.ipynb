```c++
## Teoria

Dendësia llogaritet me formulën:

\[
\rho = \frac{m}{V}
\]

ku vëllimi:

\[
V = a \cdot b \cdot h
\]

Pasiguria relative:

\[
\frac{\Delta \rho}{\rho} =
\sqrt{
\left(\frac{\Delta m}{m}\right)^2 +
\left(\frac{\Delta a}{a}\right)^2 +
\left(\frac{\Delta b}{b}\right)^2 +
\left(\frac{\Delta h}{h}\right)^2
}
\]
```

```c++
import numpy as np

# Matjet
a = 5.02   # cm
b = 3.01   # cm
h = 2.00   # cm
m = 120.2  # g

# Pasiguritë
da = 0.01
db = 0.01
dh = 0.01
dm = 0.1

# Llogaritjet
V = a * b * h
rho = m / V

# Pasiguria
rel = np.sqrt((dm/m)**2 + (da/a)**2 + (db/b)**2 + (dh/h)**2)
drho = rho * rel

print("Vëllimi =", round(V,2), "cm³")
print("Dendësia =", round(rho,2), "g/cm³")
print("Pasiguria =", round(drho,2), "g/cm³")
```

```c++
## Rezultati

Nga llogaritjet morëm:

\[
\rho \approx 3.99 \pm 0.03 \; g/cm^3
\]
```
